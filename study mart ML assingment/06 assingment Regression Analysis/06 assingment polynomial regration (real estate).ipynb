```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('real estate.csv')
```

```python
df1.head()
```

```python
df1.corr
```

```python
df1 = df1.drop(['No','X5 latitude','X6 longitude'], axis=1)
df1.head()
```

```python
df1.corr()
```

```python
df1['X1 transaction date'] = df1['X1 transaction date'].astype(str).str.extract(r'^(\d{4})')[0]
df1.head()
```

```python
df1['X1 transaction date'] = df1['X1 transaction date'].astype(int)
df1.head()
```

```python
df1.info()
```

# 1. Split the Dataset

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
```

```python
x1 = df1[['X1 transaction date','X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores']] 
y1= df1[['Y house price of unit area']]
```

```python
x1
```

```python
y1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.3, random_state=42)
```

```python
print ("(""x1_train.shape = ",x1_train.shape)
print ("(""y1_train.shape = ",y1_train.shape)
print("\n")
print ("(""x1_test.shape = ",x1_test.shape)
print ("(""y1_test.shape = ",x1_test.shape)
```

# 2. Check Correlation

```python
df1_corr = df1.corr()
df1_corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 8))
sns.heatmap(df1_corr, annot=True, fmt=".2f", cmap="RdYlGn", square=True)
plt.title("Correlation Matrix Heatmap for Real Estate")
plt.tight_layout()
plt.show()
```

# 3. Apply Regression Models ( a. Linear Regression )


```python
reg1 = LinearRegression()
reg1.fit(x1_train, y1_train)

print("Coefficients:", reg1.coef_)
print("Intercept:", reg1.intercept_)
```

```python
y1_pred = reg1.predict(x1_test)

print("Predicted values: \n", y1_pred[:5])
print("\nActual values:\n", y1_test[:5].values)
```

# 3. Apply Regression Models ( b. Polynomial Regression )


```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=4, include_bias=False)
x1_train_poly = poly.fit_transform(x1_train)
x1_test_poly = poly.transform(x1_test)
```

```python
reg_2 = LinearRegression()
reg_2.fit(x1_train_poly, y1_train)
```

```python
y1_pred_poly = reg_2.predict(x1_test_poly)
```

```python
print("Predicted values:\n", y1_pred_poly[:5])
print("\nActual values:\n", y1_test[:5].values)
```

# 4. Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

mse_linear = mean_squared_error(y1_test, y1_pred)
mse_poly = mean_squared_error(y1_test, y1_pred_poly)

r2_linear = r2_score(y1_test, y1_pred)
r2_poly = r2_score(y1_test, y1_pred_poly)
```

```python
print("Linear Regression:")
print(f"  MSE: {mse_linear:.4f}")
print(f"  R²: {r2_linear:.4f}")

print("\nPolynomial Regression:")
print(f"  MSE: {mse_poly:.4f}")
print(f"  R²: {r2_poly:.4f}")
```

# 5. Visualize Results

```python
plt.figure(figsize=(12, 5))

# Linear Regression
plt.subplot(1, 2, 1)

plt.scatter(y1_test, y1_pred, alpha=0.6, color='blue', label="Predicted")
plt.plot(y1_test, y1_test, color='red', label="Perfect Fit (y=x)")
plt.title("Linear Regression: Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()

# Polynomial Regression
plt.subplot(1, 2, 2)

plt.scatter(y1_test, y1_pred_poly, alpha=0.6, color='green', label="Predicted")
plt.plot(y1_test, y1_test, color='red', label="Perfect Fit (y=x)")
plt.title("Polynomial Regression: Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()

plt.tight_layout()
plt.show()
```
